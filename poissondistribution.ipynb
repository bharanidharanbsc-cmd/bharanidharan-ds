```python
import numpy as np
import pandas as pd
from scipy.stats import poisson as p
lam=10        
k=15
probability=p.pmf(k,lam)
probability
print("probabilty of getting 5 customers per min:",probability)
```

```python
x=np.arange(0,31)
import matplotlib.pyplot as plt
fit=p.pmf(x,lam)
plt.bar(x,fit,color="purple",label="poission PMF")
plt.bar(k,probability,color="brown",edgecolor="white",label=f"k={k}")
plt.title("Poisson Distribution")
plt.xlabel("no.of calls per min")
plt.ylabel("probability")
plt.grid()
plt.legend()
plt.show()
```

```python
s=np.arange(0,21)
import matplotlib.pyplot as plt1
plt1.bar(s,fit,color="red",label="Poission CDF")
plt1.bar(k,probability,color="yellow",edgecolor="black",label=f'k={k}')
plt1.title("poission Distribution For CDF")
plt1.xlabel("success")
plt1.ylabel("Probability of CDF")
plt1.legend()
plt1.grid(axis="y",alpha=.1)
plt1.show()
```

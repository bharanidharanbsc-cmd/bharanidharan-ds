```python
import numpy as np
import pandas as pd
data=pd.read_csv("retail_sales.csv")
data
```

```python
b_profits=data["Profit"]
b_profits
```

```python
from scipy.stats import kurtosis as k
```

```python
kurto=k(b_profits,fisher=True)
kurto1=k(b_profits,fisher=False)
mean_value=np.mean(b_profits)
max_value=max(b_profits)
min_value=min(b_profits)
print("kurtosis",kurto)
if kurto>0:
    print("heavy tail more distribution")
    if (max_value-mean_value)>(mean_value-min_value):
        print("high profit")
    else:
        print("high loss")
elif kurto<0:
    print("low tail fewer distribution")
else:
    print("normal distribution")

```

```python
import matplotlib.pyplot as plt
plt.hist(b_profits,color="green",edgecolor="black")
plt.title("profit or loss distribution")
plt.xlabel("profit values")
plt.ylabel("frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(b_profits,color="blue")
plt.title("profit or loss distribution")
plt.xlabel("profit values")
plt.ylabel("frequency")
plt.show()
```

```python
mean_value=np.mean(b_profits)
max_value=max(b_profits)
min_value=min(b_profits)
median_value=np.median(b_profits)
print("mean_value",mean_value)
print("median_value",median_value)
```

```python
import matplotlib.pyplot as plt1
```

```python
plt1.plot(b_profits,color="brown")
plt1.title("profit vs loss distribution")
plt1.xlabel("b_profits")
plt1.ylabel("frequency")
plt1.axhline(mean_value,linestyle="--",linewidth=4)
plt1.axhline(median_value,linestyle="dotted",linewidth=2)
plt1.show()
```
